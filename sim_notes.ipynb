When an order is placed we get three values
the buy, sell and stop prices. These will be 
used to construct an order object and directly
appended to the order list. This list will be 
checked for active orders on each step for status changes. On status
changes the appropriate values should be updated
and the order should be marked as completed.

- Re-calculate the account value based on the current price of held positions + the liquid value of the account [@TODO fill this parameter]

- buying power needs to be reduced (qty * cost)
- regt_buying_power to be reduced (qty * cost)
- non_marginable_buying power should be reduced (qty * cost) / 2
- Cash reduced by (qty * cost) / 2
